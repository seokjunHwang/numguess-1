# 1번 문제 
* 정리되지 않은 리스트를 정리해보세요.
# [상황]
* 아래 리스트안에는[ 한국, 독일, 미국, 일본, 이탈리아 ]의 자동차 브랜드가 정리가 되지않아 각 국가의 브랜드가 섞여있습니다.
# [필수 조건]
* 사용자에게 한국, 독일, 미국, 일본, 이탈리아 중에서만 입력을 받도록하세요.
* 각 국가를 입력받고 슬라이싱을 이용하여 입력받은 국가의 자동차 브랜드만 출력되도록 하세요.
# [선택 조건]
* List Comprehension을 이용하여 리스트를 정리해보세요.

(Comprehension을 이용할 시 굳이 슬라이싱을 고집할 필요는 없습니다.)

```
- 출력 예시 -
> 찾고자하는 국가를 입력하세요: 중국
> [ 한국, 독일, 미국, 일본, 이탈리아 ] 중 입력하세요
> 찾고자하는 국가를 입력하세요: 독일
> ['벤츠', 'BMW', '아우디']
```



````python
CarBrand = ["한국 : 현대","독일 : 벤츠","한국 : 기아","미국 : 쉐보레","이탈리아 : 람보르기니",
            "한국 : 제네시스","일본 : 혼다","미국 : 지프","미국 : 포드","이탈리아 : 페라리",
            "일본 : 인피니티", "독일 : BMW","미국 : 링컨", "이탈리아 : 마세라티","독일 : 아우디", "일본 : 도요타"]
````

````python
# 입력받을 국가만 모아둔 country 리스트 생성
country= ['한국', '독일', '미국', '일본', '이탈리아']

while True:
  input_country = input("자동차 브랜드의 국가를 입력하세요: ")
  # 입력받은 국가가 country 리스트에 없으면 사용자에게 문구를 출력하고 재입력을 받음
  if input_country not in country:
    print("[ 한국, 독일, 미국, 일본, 이탈리아 ] 중 입력하세요")
  else:
    # List Comprehension을 이용하여 new_list를 생성
    new_list = [ x.split(':')[-1] for x in CarBrand if input_country in x ]
    break
print(new_list)
````


* List Comprehension 기본 문법

`[(변수를 활용한 값) for (사용할 변수 이름) in (순회할 수 있는 값)]`

# 1번 문제 List Comprehension 부분 설명
### new_list = [ x.split(':')[-1] for x in CarBrand if input_country in x ] 에 대한 설명

1.   new_list는 List Comprehension을 사용하여 생성되며, CarBrand 리스트에서 
입력받은 input_country가 포함된 값을 추출하여 새로운 리스트를 만듦
2.   x.split(':')[-1]은 각 문자열 'x'를 ':'기준으로 분리하여, 분리된 값들을 포함하는 리스트를 생성하며, '[-1]'은 이 리스트의 마지막 요소를 가리킨다. 이렇게 하면 각 문자열 x에서 ':'이후의 문자열만 추출됨
* ex) "한국 : 현대"라는 요소를':'기준으로 분리 하면 ['한국', '현대']가 되고 리스트의 마지막 요소인 '현대'가 추가되는 것이다.
3. for x in CarBrand 부분은 CarBrand 리스트에서 각 값을 하나씩 꺼내서 x라는 변수에 할당하며,
4. if input_country in x 부분은 각 값 x에서 입력받은 input_country가 포함되어 있는 경우에만 리스트에 포함됨
5. 요약하면 'CarBrand' 리스트에서 입력받은 'input_country'가 포함된 값들 중':'이후의 문자열만을 추출하여 새로운 리스트 'new_list'를 생성하는 것이다.

````python
# List Comprehension 사용 안하고 리스트 슬라이싱을 이용
# new_list라는 빈 리스트 생성
new_list =[]
# 입력받을 국가만 모아둔 country 리스트 생성
country= ['한국', '독일', '미국', '일본', '이탈리아']

# 자동차 브랜드의 국가를 입력받는 while문
while True:
  input_country = input("자동차 브랜드의 국가를 입력하세요: ")
  # 입력받은 국가가 country 리스트에 없으면 사용자에게 문구를 출력하고 재입력을 받음
  if input_country not in country:
    print("[ 한국, 독일, 미국, 일본, 이탈리아 ] 중 입력하세요")
  else:
    # 입력받은 국가가 country 리스트에 있으면, CarBrand리스트를 확인하여
    for x in CarBrand:
     if input_country in x:
       # 해당 국가에 해당하는 브랜드를 new_list에 저장
       # 아래 코드는 new_list에 x 문자열에서 ":"을 찾은 후 
       # 뒤에 2만큼 이동한 위치부터 문자열의 끝까지 잘라내어(slice) new_list에 추가하는 코드 
       new_list.append(x[x.find(":")+2:])
    break
print(new_list)
````

# * List Comprehension의 장점
1. 코드가 간결하다.
2. 데이터베이스를 조회하여 리스트를 만들 때 많이 사용되므로, 새로운 배열을 만들때 특히 유용하다.
3. 반복문을 활용하여 새로운 배열을 만드는것 보다 처리 속도가 빠르다.

# 2번 문제
* random 모듈을 이용하여 사용자에게 입력받고, 입력받은 값을 랜덤으로 출력해보세요.
# [상황]
* 수강생 여러분은 먹고싶은것이 많아 점심메뉴로 무엇을 먹어야 할지 고르지 못하고 있습니다.
# [조건]
* 메뉴와 디저트를 입력받고, 더이상 입력을 원하지 않을시 q를 입력하여 종료하세요.
* 메뉴는 더 추가하고싶지만 디저트를 더이상 추가하고싶지 않을 상황을 고려하여 작성하세요.
* 배달을 할지 가서 먹을지 랜덤으로 추천해주는 코드를 추가하세요.
* f-string을 사용하여 print 한줄에 메뉴, 디저트, 식사 위치를 모두 출력하세요.


```
- 출력 예시 -
> 메뉴를 입력하세요 (종료는 'q' 입력): 한식
> 메뉴를 입력하세요 (종료는 'q' 입력): 중식
> 메뉴를 입력하세요 (종료는 'q' 입력): 양식
> 메뉴를 입력하세요 (종료는 'q' 입력): 일식
> 메뉴를 입력하세요 (종료는 'q' 입력): q
> 디저트를 입력하세요 (종료는 'q' 입력): 아이스크림
> 디저트를 입력하세요 (종료는 'q' 입력): 커피
> 디저트를 입력하세요 (종료는 'q' 입력): q
> 추천 메뉴: 한식
> 추천 디저트: 커피
> 식사 위치: 배달
```



````python
import random
# 메뉴와 디저트를 저장할 리스트
menu_list = []
dessert_list = []

# 메뉴 입력받는 while 문
while True:
    # 메뉴 입력
    menu = input("메뉴를 입력하세요 (종료는 'q' 입력): ")
    # 입력받은 값이 q면 while문 종료
    if menu == 'q':
        break
    # 입력받은 메뉴를 리스트에 추가
    menu_list.append(menu)

# 디저트 입력받는 while문
while True:
    # 디저트 입력
    dessert = input("디저트를 입력하세요 (종료는 'q' 입력): ")
    # 입력받은 값이 q면 while문 종료
    if dessert == 'q':
        break
    # 입력받은 디저트를 리스트에 추가
    dessert_list.append(dessert)

#메뉴와 디저트를 랜덤으로 선택
random_menu = random.choice(menu_list)
random_dessert = random.choice(dessert_list)

#배달을 할지 가서 먹을지를 랜덤으로 정하는 코드
location = ["배달", "직접"]
random_location = random.choice(location)

# 추천 메뉴, 디저트, 식사 위치 출력
# f-string을 사용해 출력
# \n을 사용해서 각각을 한줄에 출력
print(f"""
추천 메뉴: {random_menu}
추천 디저트: {random_dessert}
식사 위치: {random_location}
""")
````

# 2번 문제 random.choice 설명


```

* 시퀀스 함수는 인자로 받은 리스트, 튜플, 문자열, range에서 무작위로 하나의 원소를 뽑습니다.
* 만약, 비어있는 시퀀스 타입의 객체를 인자로 넣는다면 (ex. 리스트가 비어있다면)indexError의 예외가 발생합니다.
(indexError : 시퀀스의 인덱스가 범위를 벗어났을때 발생하는 에러)
```

```
- Refactoring을 하는 이유 -
1. 소프트웨어 설계에서 질적 향상을 위해 리펙토링을 하며, 코드 중복을 제거하고, 수정 용이성을 향상 시킴
2. 가독성 향상 
3. 버그를 찾는데 도움이 된다. 버그를 없애거나 새로운 기능을 추가하는 행위는 아니다.
4. 프로그램 개발 속도가 향상된다. 좋은 설계 기반에선, 개발 속도를 단축 할 가능성이 높아진다.
```


```
- Refactoring은 언제 할까? -
1. The Rule Of Three : 유사한 내용이 세번 이상 반복할 때, 리펙토링을 고려(꼭 세번 이상 반복되어야만 진행하는 것이 아닙니다!)
* 똑같거나 비슷한 내용이 세 번이상 작성되있을때 (무조건 진행 하는게 아니라 상황을 고려하여 리펙토링을 결정하는 것)

2. 새로운 기능을 추가할 때
* 지금 작성된 설계, 소스코드에서 새로운 기능을 추가하기 어려워 보이면, 리펙토링을 해야된다.
  이러한 상황인 경우 유지보수성이 떨어지며, 가독성 역시 좋지 않을 수 있다.

3. 코드리뷰를 할 때
* 협업하는 개발팀과 함께 코드리뷰는 질을 높일 수 있다. 그러나 인원이 너무 많은 경우에는
  비효율적이므로 소수인원으로 진행하는걸 권장한다.
  그러나 설계 리뷰에서는 많은 인원이 하여도 효과적이다.(개발 프로세스에서 상위 작업임을 고려)
```




# 문제1,2번 refactoring 진행




````python
#문제1번
# 사용자에게 국가 입력 받음
input_country = input("자동차 브랜드의 국가를 입력하세요: ")

# 입력 값이 유효한 국가가 아니면 다시 입력 받음
while input_country not in ['한국', '독일', '미국', '일본', '이탈리아']:
  input_country = input("[ 한국, 독일, 미국, 일본, 이탈리아 ] 중 입력하세요 : ")

# 입력한 국가에 해당하는 브랜드만 목록으로 새로운 리스트 생성
new_list = [ x.split(':')[-1] for x in CarBrand if input_country in x ]
print(new_list)
````



```
- Refactoring 진행 후 추가, 향상된 부분 -
1. 굳이 while True문을 사용하지 않아도 되서 while not 구문을 활용하여 코드 경량화
2. 가독성 향상
```

````python
#문제 2번
import random

# 사용자로부터 item_type에 대한 데이터를 입력 받고 items 리스트에 추가하는 함수
# input될 items : menu_list, dessert_list / item_type : '메뉴', '디저트'
def get_input(items, item_type): 
    while True:
        # item_type에 대한 데이터 입력
        item = input(f"{item_type}를 입력하세요 (종료는 'q' 입력): ")
        # 입력이 'q'일 경우 함수 종료
        if item == 'q':
            return
        # 입력된 데이터 items 리스트에 추가
        items.append(item)
        
        # 더 입력할지 묻는 부분
        cont = input(f"현재 고른 {item_type}: {items} 추가 입력 하시겠습니까?(y/n) ")
        # y 또는 n을 입력할 때까지 반복
        while cont != 'y' and cont != 'n':
            print("y 또는 n을 입력하세요.")
            cont = input(f"현재 고른 {item_type}: {items} 추가 입력 하시겠습니까?(y/n) ")
        # n일 경우 함수 종료
        if cont == 'n':
            return

# 메뉴와 디저트 데이터를 저장할 리스트
menu_list = []
dessert_list = []

# 메뉴와 디저트 데이터 입력
get_input(menu_list, "메뉴")
get_input(dessert_list, "디저트")

# 메뉴와 디저트 데이터에서 랜덤 선택
random_menu = random.choice(menu_list)
random_dessert = random.choice(dessert_list)
location = ["배달", "직접"]
random_location = random.choice(location)

print(f"""
추천 메뉴: {random_menu}
추천 디저트: {random_dessert} 
식사 위치: {random_location}
""")
````



```
- Refactoring 진행 후 추가, 향상된 부분 -
1. 서비스화를 가정하여 현재 고른 메뉴와 디저트를 확인 할 수 있게 하고 
   사용자가 추가하고자하는 item_type(음식)이 더 있는지 질문하는 코드 추가
2. 질문에 대해 "yes" or "no"만 입력 받게 하는 코드 추가
3. 반복되는 코드를 함수로 정의하여 코드의 가독성 및 유지보수성 향상
```
